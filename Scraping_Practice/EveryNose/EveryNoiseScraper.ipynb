```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://everynoise.com/artistprofile.cgi?id=1DxLCyH42yaHKGK3cl5bvG&hide=appears_on'
HEADERS = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/117.0.0.0 Safari/537.36 Avast/117.0.0.0',
    'Accept-Language': 'en-US, en;q=0.5'
}
response = requests.get(url, headers=HEADERS)
soup = BeautifulSoup(response.content,'html.parser')
```

```python
Names = []
Dates = []
Times = []
Links = []
album_box = soup.find('div', class_='albumbox', id='alltracks')
album_box_html = str(album_box)
album_box_soup = BeautifulSoup(album_box_html, 'html.parser')
songs = album_box_soup.find_all('tr', class_='trackrow')
```

```python
for song in songs:
    name = song.find('a')
    date = song.find('td', class_='date').get_text()
    time = song.find('td', class_='stat play').get_text()
    Names.append(name.get_text())
    Links.append(name.get('href'))
    Dates.append(date)
    Times.append(time)
```

```python
Data = {
    'Song Name': Names,
    'Released Date':Dates,
    'Song length' : Times,
    'Link':Links
}
```

```python
df = pd.DataFrame(Data)
df
```

```python
df.to_csv('Snehit_ki_mausi_ke_songs.csv', index=False)
```
