```python
import os
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
from urllib.request import urlretrieve

# Define the URL of the site
url = "https://www.disneystore.eu/home-and-stationery#"

# Directory to save images
output_dir = "disney_images"
os.makedirs(output_dir, exist_ok=True)

# Headers to mimic a browser visit
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}

# Send a GET request to the website
response = requests.get(url, headers=headers)
if response.status_code != 200:
    print("Failed to fetch the webpage. Status code:", response.status_code)
    exit()

# Parse the webpage content with BeautifulSoup
soup = BeautifulSoup(response.content, "html.parser")

# Find all image tags
image_tags = soup.find_all("img")

# Loop through all image tags and download the images
for idx, img_tag in enumerate(image_tags):
    # Extract the image source
    img_url = img_tag.get("src") or img_tag.get("data-src")
    if not img_url:
        continue  # Skip if there's no image source

    # Make the URL absolute
    full_url = urljoin(url, img_url)

    # Define the filename
    file_name = os.path.join(output_dir, f"image_{idx + 1}.jpg")

    # Download and save the image
    try:
        urlretrieve(full_url, file_name)
        print(f"Downloaded: {file_name}")
    except Exception as e:
        print(f"Failed to download {full_url}: {e}")

print("Image scraping complete!")
```
