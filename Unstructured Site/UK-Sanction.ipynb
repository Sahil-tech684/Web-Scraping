```python
import requests
from bs4 import BeautifulSoup

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/117.0.0.0 Safari/537.36 Avast/117.0.0.0',
    'Cookie': 'summary_id=65a393548bc62'
}

response = requests.get('https://docs.fcdo.gov.uk/docs/UK-Sanctions-List.html', headers=headers)

soup = BeautifulSoup(response.text,'html.parser')
```

```python
ids_tags = soup.select('b:contains("Unique ID:")')
div1_tags = soup.select('b:contains("Regime Name:")')
div2_tags = soup.select('div:contains("Regime Name:")')

IDS = []
Div1 = []
Div2 = []
Div3 = []
Div4 = []
Div5 = []
Div6 = []
Div7 = []
Div8 = []

for ids_tag in ids_tags:
    next_sibling = ids_tag.find_next_sibling()
    if next_sibling:
        content = next_sibling.text.strip()
        IDS.append(content)

for div1_tag in div1_tags:
    siblings = div1_tag.find_next_siblings()
    content = ' '.join([sibling.text.strip() for sibling in siblings])
    Div1.append(content)

for div2_tag in div2_tags:
    next_siblings = div2_tag.find_all_next('div')
    content = next_siblings[0].text.strip() if next_siblings else ''
    content1 = next_siblings[1].text.strip() if next_siblings else ''
    content2 = next_siblings[2].text.strip() if next_siblings else ''
    content3 = next_siblings[3].text.strip() if next_siblings else ''
    content4 = next_siblings[4].text.strip() if next_siblings else ''
    content5 = next_siblings[5].text.strip() if next_siblings else ''
    content6 = next_siblings[6].text.strip() if next_siblings else ''
    Div2.append(content) 
    Div3.append(content1)
    Div4.append(content2)
    Div5.append(content3)
    Div6.append(content4)
    Div7.append(content5)
    Div8.append(content6)

```
