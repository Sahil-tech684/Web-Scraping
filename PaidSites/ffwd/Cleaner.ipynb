```python
import pandas as pd

df = pd.read_csv('ffwdData.csv')
df
```

```python
founders_split = df['FoundersLinkedIn'].str.split(', ', expand=True)

# Renaming the new columns
founders_split.columns = [f'Founder{i+1}LinkedIn' for i in range(founders_split.shape[1])]

# Concatenating the new columns to the original DataFrame
df = pd.concat([df, founders_split], axis=1)

# Dropping the original FoundersName column
df = df.drop(['FoundersLinkedIn'], axis=1)
df
```

```python
df.to_csv('ffwdCleanedData.csv',index=False)
```
